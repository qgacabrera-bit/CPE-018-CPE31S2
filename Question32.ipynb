```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
circle = cv2.imread('/content/drive/MyDrive/3rd Year Subs/circles.png')
gray_img = cv2.cvtColor(circle, cv2.COLOR_BGR2GRAY)
img = cv2.medianBlur(gray_img, 35)
cimg = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)
circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,1,100,param1=100,param2=30,minRadius=0,maxRadius=0)
circles = np.uint16(np.around(circles))

for i in circles[0,:]:
  cv2.circle(circle,(i[0],i[1]),i[2],(0,0,255),2)
cv2_imshow(circle)
cv2.imwrite('/content/drive/MyDrive/3rd Year Subs/circles_detected.png',circle)
```
